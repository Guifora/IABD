## EJERCICIO 1

Dibuja 100 puntos aleatorios (entre 0 y 100 para ambas coordenadas) con colores y tamaños aleatorios y una transparencia de 0.5.

Tendrás que crear 4 arrays aleatorios de 100 posiciones: uno para las coordenadas x, otro para y, otro para el color de cada punto y un último para el tamaño de cada punto. Para que los puntos no se vean muy pequeños, multiplica el array de tamaños por 10.

Configura el tamaño de la figura a 10 x 10. Para ello, investiga el método .figure() de plt.

Por último, selecciona los colores combinando el array aleatorio y el mapa de color llamado hsv y recuerda mostrar la colorbar.

````python
# Generamos un conjunto de 100 valores aleatorios uniformemente distribuidos entre 0 y 100 para la coordenada x.
x = np.random.uniform(0, 100, size=100)

# Generamos un conjunto de 100 valores aleatorios uniformemente distribuidos entre 0 y 100 para la coordenada y.
y = np.random.uniform(0, 100, size=100)

# Generamos un conjunto de 100 valores aleatorios entre 0 y 1 para el color de los puntos en el gráfico.
color = np.random.rand(100)

# Generamos un conjunto de 100 valores enteros aleatorios entre 0 y 10, y los multiplicamos por 10 para definir el tamaño de los puntos.
size = np.random.randint(0, 10, size=100) * 10

# Creamos una figura con un tamaño de 10x10 pulgadas.
plt.figure(figsize=(10, 10))

# Creamos un gráfico de dispersión (scatter plot) con las siguientes propiedades:
# - x: coordenadas en el eje horizontal.
# - y: coordenadas en el eje vertical.
# - c: colores de los puntos, definidos por el arreglo 'color'.
# - cmap: mapa de colores utilizado, en este caso "hsv" (espectro de colores).
# - alpha: nivel de transparencia de los puntos (0.5 significa un 50% de opacidad).
# - s: tamaños de los puntos, definidos por el arreglo 'size'.
plt.scatter(x=x, y=y, c=color, cmap="hsv", alpha=0.5, s=size)

# Añadimos un título al gráfico.
plt.title("Mapa de color plasma")

# Añadimos una barra de color al gráfico, que indica el rango de valores del color utilizado.
plt.colorbar()

# Mostramos el gráfico en pantalla.
plt.show()
````

## EJERCICIO 2

Crea una función que dado un color en forma RGB lo transforme en su forma hexadecimal.

````python
def rgb_a_hexadecimal(r, g, b):
    """
    Convierte un color RGB a su representación hexadecimal.

    Parámetros:
    r, g, b (int): Valores de rojo, verde y azul (entre 0 y 255).

    Retorna:
    str: El color en formato hexadecimal (ejemplo: #RRGGBB).
    """
    # Validar que los valores estén en el rango 0-255
    if (0 <= r <= 255) and (0 <= g <= 255) and (0 <= b <= 255):
        return f"#{r:02X}{g:02X}{b:02X}"
    else:
        return "Valores RGB fuera de rango. Deben estar entre 0 y 255."

# Ejemplo de uso
color_rgb = (255, 99, 71)  # Tomate
color_hex = rgb_a_hexadecimal(*color_rgb)
print(f"El color {color_rgb} en hexadecimal es: {color_hex}")
````

````python
# Generamos un conjunto de valores para la variable `x`, espaciados uniformemente entre -5π y 5π.
# `np.linspace(-5*np.pi, 5*np.pi, 100)` crea un arreglo de 100 puntos desde -5π hasta 5π.
x = np.linspace(-5 * np.pi, 5 * np.pi, 100)

# Definimos la primera función `f1` como sin(x) / x, que corresponde a la función sinc.
f1 = np.sin(x) / x

# Definimos la segunda función `f2` como 2 * (sin(x) / x), que es la función `f1` multiplicada por 2.
f2 = 2 * (np.sin(x) / x)

# Definimos la tercera función `f3` como -sin(x) / x, que es la función `f1` invertida.
f3 = -np.sin(x) / x

# Creamos una figura de tamaño 8x6 pulgadas para los gráficos.
plt.figure(figsize=(8, 6))

# Dibujamos la primera función `f1` con la variable `x` en el gráfico, utilizando color naranja (#f06741).
plt.plot(x, f1, color="#f06741")

# Dibujamos la segunda función `f2` con color verde (#a4f041).
plt.plot(x, f2, color="#a4f041")

# Dibujamos la tercera función `f3` con color morado (#8d41f0).
plt.plot(x, f3, color="#8d41f0")

# Mostramos el gráfico con las tres curvas.
plt.show()
````

````python
# Importamos la biblioteca NetworkX para la creación y manipulación de grafos.
import networkx as nx

# Creamos un grafo vacío llamado G.
G = nx.Graph()

# Añadimos nodos al grafo utilizando los caracteres de la cadena "abcde".
# Esto crea los nodos "a", "b", "c", "d" y "e" en el grafo.
G.add_nodes_from("abcde")

# Añadimos aristas ponderadas (con peso) al grafo. Cada tupla tiene la forma (nodo1, nodo2, peso).
# Representa las conexiones entre los nodos con un peso asociado:
# - "a" conectado a "b" con peso 1
# - "a" conectado a "c" con peso 2
# - "a" conectado a "d" con peso 2
# - "a" conectado a "e" con peso 5
# - "b" conectado a "e" con peso 4
# - "c" conectado a "d" con peso 3
# - "d" conectado a "e" con peso 1
G.add_weighted_edges_from([("a", "b", 1),
                           ("a", "c", 2),
                           ("a", "d", 2),
                           ("a", "e", 5),
                           ("b", "e", 4),
                           ("c", "d", 3),
                           ("d", "e", 1)])

# Creamos una figura y un eje para dibujar el grafo con Matplotlib.
fig, ax = plt.subplots()

# Dibujamos el grafo con NetworkX, usando diferentes propiedades visuales para los nodos y aristas.
nx.draw(G, ax=ax,
        with_labels=True,              # Muestra las etiquetas (nombres de los nodos).
        font_weight="bold",            # Estilo de fuente en negrita para las etiquetas de los nodos.
        font_color="white",            # Color del texto de las etiquetas de los nodos.
        node_color="#ad80fa",          # Color de los nodos (un tono morado claro).
        edge_color="#b1ddf5")          # Color de las aristas (un tono azul celeste).

# Muestra el grafo en la ventana de gráficos.
plt.show()
````

## EJERCICIO 7

Considera el grafo del ejercicio anterior. El tamaño de los nodos debe ser 350 por el grado del nodo. El tamaño de la figura global debe ser de 8 x 6.

````python
# Importamos la biblioteca NetworkX para trabajar con grafos.
import networkx as nx

# Creamos un grafo vacío llamado G.
G = nx.Graph()

# Añadimos nodos al grafo usando los caracteres de la cadena "abcde".
# Esto crea los nodos "a", "b", "c", "d", y "e".
G.add_nodes_from("abcde")

# Añadimos aristas ponderadas (con peso) al grafo.
# Cada tupla tiene la forma (nodo1, nodo2, peso), representando la conexión entre nodos.
G.add_weighted_edges_from([("a", "b", 1),  # Conexión entre "a" y "b" con un peso de 1.
                           ("a", "c", 2),  # Conexión entre "a" y "c" con un peso de 2.
                           ("a", "d", 2),  # Conexión entre "a" y "d" con un peso de 2.
                           ("a", "e", 5),  # Conexión entre "a" y "e" con un peso de 5.
                           ("b", "e", 4),  # Conexión entre "b" y "e" con un peso de 4.
                           ("c", "d", 3),  # Conexión entre "c" y "d" con un peso de 3.
                           ("d", "e", 1)]) # Conexión entre "d" y "e" con un peso de 1.

# Creamos una figura y un eje para dibujar el grafo usando Matplotlib.
fig, ax = plt.subplots()

# Obtenemos un diccionario con el grado de cada nodo (número de conexiones que tiene).
# Por ejemplo, {'a': 4, 'b': 2, 'c': 2, 'd': 3, 'e': 3}.
dicc = dict(G.degree())

# Calculamos el tamaño de los nodos en función del número de conexiones (grado).
# Multiplicamos cada grado por 350 para hacer los nodos más grandes y distinguibles.
grosor_nodos = [x * 350 for x in dicc.values()]

# Dibujamos el grafo con NetworkX usando varias propiedades de estilo.
nx.draw(G, ax=ax,
        with_labels=True,              # Muestra las etiquetas de los nodos.
        font_weight="bold",            # El texto de las etiquetas de los nodos estará en negrita.
        font_color="white",            # El color del texto de las etiquetas será blanco.
        node_color="#ad80fa",          # Color de los nodos (un tono morado claro).
        edge_color="#b1ddf5",          # Color de las aristas (un tono azul celeste).
        node_size=grosor_nodos)        # Tamaño de los nodos ajustado según el número de conexiones.

# Ajusta el tamaño de la figura a 8x6 pulgadas para mejorar la visualización.
plt.figure(figsize=(8, 6))

# Muestra el grafo en la ventana de gráficos.
plt.show()
````

## EJERCICIO 8

Considera el grafo del ejercicio anterior. El grosor de las aristas debe ir acorde al peso de cada arista.

````python
# Importamos la biblioteca networkx para trabajar con grafos.
import networkx as nx
# Creamos un nuevo grafo vacío G.
G = nx.Graph()

# Añadimos nodos al grafo usando los caracteres de la cadena "abcde".
# Cada letra se convierte en un nodo del grafo.
G.add_nodes_from("abcde")

# Añadimos aristas ponderadas al grafo (con pesos). Cada tupla representa (nodo1, nodo2, peso).
G.add_weighted_edges_from([("a", "b", 1),  # Arista entre "a" y "b" con peso 1.
                           ("a", "c", 2),  # Arista entre "a" y "c" con peso 2.
                           ("a", "d", 2),  # Arista entre "a" y "d" con peso 2.
                           ("a", "e", 5),  # Arista entre "a" y "e" con peso 5.
                           ("b", "e", 4),  # Arista entre "b" y "e" con peso 4.
                           ("c", "d", 3),  # Arista entre "c" y "d" con peso 3.
                           ("d", "e", 1)]) # Arista entre "d" y "e" con peso 1.

# Crea una figura y un eje usando Matplotlib.
fig, ax = plt.subplots()

# Crea un diccionario con el grado de cada nodo.
# G.degree() devuelve el número de conexiones (grado) de cada nodo en el grafo.
dicc = dict(G.degree())

# Calcula el tamaño de los nodos en función del grado.
# Multiplica el grado por 350 para aumentar la diferencia visual entre tamaños.
grosor_nodos = [x * 350 for x in dicc.values()]

# Obtiene un diccionario con los pesos de cada arista.
# nx.get_edge_attributes(G, "weight") devuelve los pesos de las aristas como un diccionario {("a", "b"): 1, ...}.
dicc2 = nx.get_edge_attributes(G, "weight")

# Crea una lista con el grosor de las aristas basada en los valores de peso.
# Los pesos de las aristas determinan el grosor de las líneas.
grosor_aristas = [i for i in dicc2.values()]

# Dibuja el grafo usando NetworkX con varias propiedades estilísticas.
nx.draw(G, ax=ax,
        with_labels=True,              # Muestra las etiquetas (nombres de nodos).
        font_weight="bold",            # Fuente en negrita para los nombres de los nodos.
        font_color="white",            # Color de las etiquetas de los nodos.
        node_color="#ad80fa",          # Color de los nodos (tono morado claro).
        edge_color="#b1ddf5",          # Color de las aristas (tono celeste).
        node_size=grosor_nodos,        # Tamaño de los nodos basado en su grado.
        width=grosor_aristas)          # Grosor de las aristas basado en sus pesos.

# Ajusta el tamaño de la figura para que sea más grande (8x6 pulgadas).
plt.figure(figsize=(8, 6))

# Muestra la figura generada.
plt.show()
````

Investiga el método .text() de plt para añadir texto a un plot.
