Del dataframe points, muestra las observaciones de la forma: “El punto ({x}, {y}) {no/sí} pertenece al primer cuadrante”.

````python
# Iteramos sobre cada fila del DataFrame 'points' utilizando el método iterrows().
# 'id' contiene el índice de la fila y 'fila' contiene los datos de la fila como una serie de pandas.
for id, fila in points.iterrows():
    # Comprobamos si el valor de 'x' de la fila es menor que -0.953741
    # Y si el valor de 'y' de la fila es menor que -0.517294
    if (fila['x'] < -0.953741) & (fila['y'] < -0.517294):
        # Si ambas condiciones son verdaderas, se imprime que el punto pertenece al primer cuadrante.
        # En este caso, se usa "si" como parte de la cadena de salida.
        print("El punto ({}, {}) {} pertenece al primer cuadrante".format(fila['x'], fila['y'], "si"))
    else:
        # Si alguna de las condiciones no es verdadera, se imprime que el punto no pertenece al primer cuadrante.
        # En este caso, se usa "no" como parte de la cadena de salida.
        print("El punto ({}, {}) {} pertenece al primer cuadrante".format(fila['x'], fila['y'], "no"))
````

## EJERCICIO 6

Crea un dataframe con 10 filas y 4 columnas. La primera columna se llamará word; la segunda, length; la tercera, vowels; y la última, consonants. La columna words contendrá las siguientes 10 palabras: “euro”, “diez” “algas”, “broma”, “cicuta”, “fatiga”, “nachos”, “jadeos”, “hazañas”, “boutique”. Las columnas length, vowels y consonants contendrán, respectivamente, la longitud, el total de vocales y el total de consonantes (añade las columnas calculando su valor por código). Guarda el dataframe en la variable words.

````python
# Definimos una función para contar las vocales en una palabra.
def contarvoc(palabra):
    # Inicializamos el contador de vocales en 0.
    cont = 0
    # Definimos una lista de las vocales.
    voc = ["a", "e", "i", "o", "u"]
    # Iteramos sobre cada carácter en la palabra.
    for char in palabra:
        # Si el carácter está en la lista de vocales, incrementamos el contador.
        if char in voc:
            cont += 1
    # Devolvemos el total de vocales encontradas.
    return cont

# Definimos una función para contar las consonantes en una palabra.
def contarcons(palabra):
    # Inicializamos el contador de consonantes en 0.
    cont = 0
    # Definimos una lista de las vocales.
    voc = ["a", "e", "i", "o", "u"]
    # Iteramos sobre cada carácter en la palabra.
    for char in palabra:
        # Si el carácter no está en la lista de vocales, incrementamos el contador.
        if char not in voc:
            cont += 1
    # Devolvemos el total de consonantes encontradas.
    return cont

# Lista de palabras a analizar.
valores = ["euro", "diez", "algas", "broma", "cicuta", "fatiga", "nachos", "jadeos", "hazañas", "boutique"]

# Creamos un diccionario para almacenar los datos de cada palabra:
# - "words": la lista de palabras.
# - "length": la longitud de cada palabra (calculada usando una lista por comprensión).
# - "vowels": el número de vocales en cada palabra (usando la función contarvoc).
# - "consonants": el número de consonantes en cada palabra (usando la función contarcons).
dicc = {
    "words": ["euro", "diez", "algas", "broma", "cicuta", "fatiga", "nachos", "jadeos", "hazañas", "boutique"],
    "length": [len(i) for i in valores],
    "vowels": [contarvoc(i) for i in valores],
    "consonants": [contarcons(i) for i in valores]
}

# Creamos un DataFrame de pandas a partir del diccionario.
words = pd.DataFrame(dicc)

# Mostramos el DataFrame resultante que contiene las palabras, su longitud, el número de vocales y consonantes.
words
````

## EJERCICIO 7

Muestra las 10 observaciones de words con el formato “La palabra {word} tiene {length} letras, de las cuales {vowels} son vocales y {consonants} consonantes”.

````python
# Iteramos sobre cada fila del DataFrame 'words' utilizando el método itertuples().
# 'id' es el índice de la fila, y 'word', 'len', 'vow', 'cons' son las columnas del DataFrame
# que contienen la palabra, su longitud, el número de vocales y el número de consonantes, respectivamente.
for id, word, len, vow, cons in words.itertuples():
    # Imprimimos un mensaje que muestra la información de cada palabra.
    # Utilizamos el método format() para insertar los valores de la palabra, su longitud,
    # el número de vocales y consonantes en la cadena de texto que se imprimirá.
    print("La palabra {} tiene {} letras, de las cuales {} son vocales y {} consonantes".format(word, len, vow, cons))
````

## EJERCICIO 8

Muestra aquellas observaciones de words que tienen el mismo número de vocales y consonantes.

````python
# Filtramos el DataFrame 'words' para obtener solo aquellas filas donde el número de vocales es igual al número de consonantes.
# La condición 'words.vowels == words.consonants' crea una serie booleana que indica si el número de vocales en cada palabra es igual al número de consonantes.
# El resultado será un nuevo DataFrame que contiene solo las palabras que cumplen con esta condición.
resultado = words[words.vowels == words.consonants]

# Mostramos el resultado filtrado.
print(resultado)
````
